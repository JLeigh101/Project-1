```python
import os
import numpy as np
import pandas as pd
```

```python
weekly_2022_12_29 = "regional-us-weekly-2022-12-29.csv"
weekly_2023_01_05 = "regional-us-weekly-2023-01-05.csv"
weekly_2023_01_12 = "regional-us-weekly-2023-01-12.csv"
weekly_2023_01_19 = "regional-us-weekly-2023-01-19.csv"
weekly_2023_01_26 = "regional-us-weekly-2023-01-26.csv"
```

```python
weekly_2022_12_29_df = pd.read_csv(weekly_2022_12_29)
weekly_2023_01_05_df = pd.read_csv(weekly_2023_01_05)
weekly_2023_01_12_df = pd.read_csv(weekly_2023_01_12)
weekly_2023_01_19_df = pd.read_csv(weekly_2023_01_19)
weekly_2023_01_26_df = pd.read_csv(weekly_2023_01_26)
```

```python
weekly_2023_01_05_df['chart']='2023_01_05'
weekly_2023_01_12_df['chart']='2023_01_12'
weekly_2023_01_19_df['chart']='2023_01_19'
weekly_2023_01_26_df['chart']='2023_01_26'
```

```python
concat_df = pd.concat([weekly_2022_12_29_df, weekly_2023_01_05_df, weekly_2023_01_12_df, weekly_2023_01_19_df, weekly_2023_01_26_df])
```

```python
concat_df.head()
```

```python
bychart = concat_df.groupby('chart')
bychart.head()
```
